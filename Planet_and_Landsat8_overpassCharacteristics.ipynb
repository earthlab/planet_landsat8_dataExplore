```python
## convert the metadata into a data frame
acq_df = pd.DataFrame.from_records(planet_acq)
```

```python
## convert the metadata into a data frame
id_df = pd.DataFrame.from_records(planet_system_ids)
```

```python
meta_df = acq_df.join(id_df)
meta_df
```

```python
for k in meta_df.keys():
    print('{} dtype: {}, {}'.format(k, meta_df[k].dtype, meta_df[k].dtype is not np.float64))
    

for k in meta_df.keys():
    if meta_df[k].dtype is not np.dtype('float64'):
        # plot by value_counts()
        meta_df[k].value_counts().plot(kind='bar', title=k)
        plt.show()
        
    else:
        # plot histogram
        meta_df[k].plot(kind='hist', title=k, bins=20)
        plt.show()
```

```python
meta_df['serialID'].value_counts()
```

```python
## group by serialID
id_group = meta_df.groupby('serialID')
id_group.groups
for name,group in id_group:
    print ('SerialID: '+ name)
    print (group)
    
    group.plot(kind='bar', title=name, figsize=(15,10))
    plt.show()
    print('\n')

```

```python
## group by acquisitionDate
id_group = meta_df.groupby('acquisitionDate')
id_group.groups
for name,group in id_group:
    print ('Acquisition Date: '+ name)
    print (group)
    
    group.plot(kind='bar', title=name, figsize=(15,10))
    plt.show()
    print('\n')
```
